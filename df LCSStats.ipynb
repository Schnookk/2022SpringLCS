```python
df['KDA'] = (df['kills'] + df['assists']) / df['deaths']
```

```python
df_teams = df[df['participantid'] >= 100]
df_teams.head()
```

```python
df_players = df[df['participantid'] < 100]
df_players[:11]
```

```python
df_teams.sort_values(by=['KDA'], ascending=False)
```

```python
df_Bjergsen = df[df['playername'] == 'Bjergsen'] #new dataframe for analyzing Bjergsen's stats
df_Bjergsen #display Bjergsen dataframe
```

```python
df_team_kda_wins = df_teams.groupby('teamname').agg({'result':'sum', 'KDA':'mean'}).rename(columns={'result':'Wins'}) #new dataframe for team kda, grouped by team name
df_team_kda_wins2 = df_team_kda_wins.sort_values(by='Wins', ascending=False).round(2) #sort dataframe descending, round for readability 
df_team_kda_wins2
```

```python
df_team_kda_wins2.plot()
```

```python
df_players_best = df_players.groupby('playername').agg({'cspm':'mean', 'kills':'mean'}).rename(columns={'cspm':'CS/min', 'kills':'Avg Kills'}).sort_values(by='CS/min', ascending=False).round(2) #new dataframe for player stats
df_players_best[:15]
```

```python
df_players_best[:45].plot()
```
